```python
# Imports 
import os
import xml.etree.ElementTree as et
import pandas as pd
from itertools import chain

import numpy as np
from sklearn.manifold import TSNE
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.decomposition import NMF
import random
random.seed(13)


#visualisation stuff
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
import matplotlib
%matplotlib inline
```

```python
# Load data 
PATH = "../DS4D-Trade-Agreement-Project/xml/"
extension = ".xml"

def load_order(PATH, extension):
    """
    Custom function for loading and sorting the xml files from the TOTA dataset.
    Returns an array of filenames sorted by the number in the file name.
    """
    files = []
    for filename in os.listdir(PATH):
        if filename == ".ipynb_checkpoints":
            continue
        files.append(filename.split(".")[0]) # remove .xml
    files.sort(key=lambda x:int(x.split("_")[1])) # sort files based on file number 

    files = [file + extension for file in files]
#     print(files)
    return files

files = load_order(PATH, extension)
```

```python
def xml2list(file):
    """
    Custom function for parsing and outputing the data into arrays.
    Returns two array's the extracted column names and content for each column.
    """
    parsed_xml = et.parse(file)

    root = parsed_xml.getroot()

    meta_column_names = []
    meta_column_content = []

    for item in root[0]:
        grandchild = item.findall(".//")
        if grandchild:
            grandstore = []
            #print(item.tag)
            meta_column_names.append(item.tag)
            for x in grandchild:
                grandstore.append(int(x.attrib.get('n')))
                #print(x.tag +":", x.attrib)
            meta_column_content.append(grandstore)

        else:
            #print(item.tag +":", item.text)
            meta_column_names.append(item.tag)
            meta_column_content.append(item.text)
            
    return meta_column_names, meta_column_content
    
    
def xml2list_text(file):
    """
    Custom function for parsing and outputing the data into arrays.
    Returns two array's the extracted column names and content for each column.
    """

    parsed_xml = et.parse(file)
    root = parsed_xml.getroot()
    chapters = []
    article_store = []
    
    treaty_text = ''
    
    for item in root[1]:
        chapter_name = item.attrib.get('name')
#         print("chapter: " + item.attrib.get('name'))
        chapters.append(chapter_name)
        if chapter_name == None:
            continue
        treaty_text += chapter_name + "\n"
        
        articles = item.findall("article")
        for article in articles:
#             print(article.attrib.get('name'))
            art_no = article.attrib.get('number')
            art_text =  article.text
            article_store.append(art_no)
            if art_no == None:
                continue
#             print(art_no)
            treaty_text += art_no + "\n"
            treaty_text += art_text + "\n"
        
            
    return len(chapters), len(article_store), treaty_text
```

```python
textdf = pd.DataFrame()

chapter_numbers = []
article_numbers = []

for i in range(0, len(files)):
    chapter_no, article_no, text = xml2list_text(PATH + files[i])
    chapter_numbers.append(chapter_no)
    article_numbers.append(article_no)

textdf["chapter_number"] = chapter_numbers
textdf["article_number"] = article_numbers
```

```python
dfcols, null = xml2list(PATH + files[3])
totadf = pd.DataFrame(columns=dfcols)

for i in range(0, len(files)):
    dfcolstmp, datatmp = xml2list(PATH + files[i])
    tmp_df = pd.DataFrame([datatmp], columns=dfcolstmp)
    tmp_df = tmp_df.loc[:,~tmp_df.columns.duplicated()] #very important! removes duplicate columns note that it removes the multiple "source" columnsns
    totadf = totadf.loc[:,~totadf.columns.duplicated()]#very important! removes duplicate columns note that it removes the multiple "source" columnsns 
    totadf = pd.concat([totadf,tmp_df], axis=0, sort=False)
totadf = totadf.reset_index(drop=True)
totadf = pd.concat([totadf, textdf], axis=1)
```

## Non-Negative Matrix Factorization (NMF)
Here I perform the document clustering using the NMF algortithm. This proccess is entierly unsupervised (which is nice) as we can just feed it data and it'll happily try to do something with it. A downside to this however, is that it'll create clusters based on the data but not based on our idea of how the data is structured - what this means is that it might return 5 different topics that have nothing to do with the type of agreement but based on other defining characteristics - or even some combination of characteristics. As such, we have to then get a feel for what each cluster reperesnts by hand. I have done this two different ways - the first is to output the top hundred most representative words for each cluster (as you can see below for this data it is quite difficult to differentiate clusters based on the top 100 words alone). The second apprach is to output the top three or so treaties that are most representative of the cluster. We can then either read the treaties ourselves to get a feel for the different cluster categories or in this case since we have access to metadata regarding the treaties we can compare the clusters based on their "name", "treaty type" and "region". This approach is very insightful for this dataset as the metadata (we hope) acurately represents the contents of the treaty. 

```python
# Parameters that I tweaked - could probably spend a whole week trying to find the best combination - but this selection seems to work just fine
n_features = 1000
n_topics = 5 # 5 because I want it to try and cluster based on the 5 treaty types
n_top_words = 100


def print_top_words(model, feature_names, n_top_words):
    for topic_idx, topic in enumerate(model.components_):
        print("\nTopic #%d:" % topic_idx)
        print(" ".join([feature_names[i]
                        for i in topic.argsort()[:-n_top_words - 1:-1]]))
    print()
#I convert a collection of raw documents into a matrix of TF-IDF features
tfidf_vectorizer = TfidfVectorizer(max_df=0.95, min_df=2,max_features=n_features,stop_words='english')
#retearns a term-document matrix
tfidf = tfidf_vectorizer.fit_transform(totadf_emb['treaty_text'])
# This is where the non negative matrix factorization happens - in short it is a form of dimensionality reduction
nmf = NMF(n_components=n_topics, random_state=0,alpha=.1, l1_ratio=.5).fit(tfidf)

print("Topics found via NMF:")
# need to print the words most related to the seperate topics - quite difficult to differentiate by hand. 
tfidf_feature_names = tfidf_vectorizer.get_feature_names()
print_top_words(nmf, tfidf_feature_names, n_top_words)
```

```python
nmf_embedding = nmf.transform(tfidf)
nmf_embedding = (nmf_embedding - nmf_embedding.mean(axis=0))/nmf_embedding.std(axis=0)
```

Below I return the most representative documents for the different topics (they are classified by their treaty identifier). I do this because as you can probably see it is quite difficult to label the above topics based puerly on the text. So I have returned the most representative treaties for each cluster which I can then at a later point read to get a better idea of what each cluster represents. However, by quickly looking at the most representative name, region and type for each cluster we can see that the NMF algorithm has been mostly succesful in clustering the different types of agreements - which is quite surprising actually. Looking at the regions there might also be a certain amount of clustering based on region.  

```python
# return the most representative documents for the different topics (they are classified by their treaty identifier)
top_idx = np.argsort(nmf_embedding, axis=0)[-3:]

count = 0
for idxs in top_idx.T:
    print("\nTopic {}:".format(count))
    for idx in idxs:
        print(totadf.iloc[idx][['name', 'region', 'type']])
    count += 1
```

## t-distributed Stochastic Neighbor Embedding
Here I use tsne to visualize the high dimensional data created by the nmf. I did some reading as to what the typical parameters are as well as what to look out for when using this technique - quite an interesting read if you have no knowledge of t-sne and how it behaves: [misread-t-sne](https://distill.pub/2016/misread-tsne/).

```python
tsne = TSNE(n_iter=5000, random_state=3211,learning_rate=100, perplexity=60) # run t-sne with the various parameters
tsne_embedding = tsne.fit_transform(nmf_embedding) 
tsne_embedding = pd.DataFrame(tsne_embedding,columns=['x','y']) # turn the results into a dataframe
tsne_embedding['hue'] = nmf_embedding.argmax(axis=1) # attach a hue column so that we can later color each cluster differently
tsne_embedding['year'] = totadf['year_signed'] # attach a year column so that we can see the evolution of the treaties over time. 
```

**Trade agreemeents clustered by topic**

I have used a combination of the non-negative matrix factorization (NMF) and t-Distributed Stochastic Neighbor Embedding (t-sne) algorithms to extract features and cluster the trade agreements as well as create a 2D-embedding representation of the relationships between the different trade agreements. NMF was used as a feature extractor for the trade agreements, it allowed me to decompose the text into a predefined number of features and then differentiate these features into a predetermined number of topics. Essentialy, it allows me to cluster the different trade agreements. A downside to this approach is that clusters are not labeled and can be hard to interpret - especialy in this case when the agreements are only subtly different (for a non-domain expert). Furthermore, NMF produces a high dimensional representation of the various trade agreements, making it hard to visualize. To this end I use the t-sne dimensionality reduction algorithm to convert the high dimensional embedding into a two dimensional one - one that we can visualise. I decided to cluster the documents based on 5 topics as this closely relates to the number of different types of agreements, however as you can see, the topics are not simply related to topic alone but also region and most likely other variables. The fact that region has an effect on a treaty type is in itself interesting (in the sense that a particulary treaty type is different in different regions).
